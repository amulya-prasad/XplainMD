## Introduction and Setup

**What is PrimeKG?**

PrimeKG is a precision medicine knowledge graph curated by Harvard PhD students that integrates drugs, diseases, phenotypes, proteins, and other medical entities into a single, structured graph. Each node represents a biomedical concept, and each edge represents a known or inferred relationship between them.

You can find more information about it here: https://github.com/mims-harvard/PrimeKG

## Section 2: Basic Graph Exploration
Before diving into complex graph algorithms or predictive modeling, it’s important to understand the basic structure and composition of the graph.

This section gives a high-level overview of:

- How many nodes and edges exist

- What types of entities are represented (e.g., drugs, diseases, proteins)

- What types of relationships connect them (e.g., drug–effect, protein–protein)

This foundational understanding helps set the stage for more advanced analyses and ensures that we’re working with the right data for our downstream tasks.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx

# Plot style
sns.set(style="whitegrid")

# ----- NODE TYPE DISTRIBUTION -----
x_node_types = filtered_df[['x_id', 'x_type']].rename(columns={'x_id': 'node_id', 'x_type': 'node_type'})
y_node_types = filtered_df[['y_id', 'y_type']].rename(columns={'y_id': 'node_id', 'y_type': 'node_type'})
all_nodes = pd.concat([x_node_types, y_node_types]).drop_duplicates()
node_type_counts = all_nodes['node_type'].value_counts().reset_index()
node_type_counts.columns = ['Node Type', 'Count']

# ----- EDGE TYPE DISTRIBUTION -----
edge_type_counts = filtered_df['relation'].value_counts().reset_index()
edge_type_counts.columns = ['Edge Type', 'Count']

# ----- PLOTTING NODE + EDGE TYPE DISTRIBUTIONS -----
fig, axs = plt.subplots(1, 2, figsize=(14, 5))

sns.barplot(data=node_type_counts, x='Node Type', y='Count', hue='Node Type', ax=axs[0], palette='coolwarm', legend=False)
axs[0].set_title("Node Type Distribution")
axs[0].tick_params(axis='x', rotation=45)

sns.barplot(data=edge_type_counts, x='Edge Type', y='Count', hue='Edge Type', ax=axs[1], palette='Set2', legend=False)
axs[1].set_title("Edge Type Distribution")
axs[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# ----- BUILD GRAPH -----
G = nx.from_pandas_edgelist(filtered_df, source='x_id', target='y_id', create_using=nx.Graph())

# Add node attributes (type + name) for x and y
for _, row in filtered_df.iterrows():
    G.nodes[row['x_id']]['type'] = row['x_type']
    G.nodes[row['x_id']]['name'] = row['x_name']
    G.nodes[row['y_id']]['type'] = row['y_type']
    G.nodes[row['y_id']]['name'] = row['y_name']

# ----- TOP-K NODES BY DEGREE -----
top_k = 15
degree_dict = dict(G.degree())
top_degree_nodes = sorted(degree_dict.items(), key=lambda x: x[1], reverse=True)[:top_k]

print(f"\n Top {top_k} Nodes by Degree:\n")
print(f"{'Node ID':<10} {'Degree':<10} {'Name':<30} {'Type'}")
print("-" * 70)
for node_id, degree in top_degree_nodes:
    name = G.nodes[node_id].get('name', 'N/A')
    node_type = G.nodes[node_id].get('type', 'N/A')
    print(f"{node_id:<10} {degree:<10} {name:<30} {node_type}")

# Prepare data for plotting
labels = []
degrees = []
colors = []

for node_id, degree in top_degree_nodes:
    name = G.nodes[node_id].get('name', str(node_id))
    node_type = G.nodes[node_id].get('type', 'unknown')
    label = f"{name} ({node_type})"
    labels.append(label)
    degrees.append(degree)
    
    # Color by type
    if "gene" in node_type:
        colors.append("skyblue")
    elif "disease" in node_type:
        colors.append("salmon")
    elif "phenotype" in node_type:
        colors.append("orchid")
    else:
        colors.append("gray")

# Plot
plt.figure(figsize=(10, 7))
plt.barh(labels[::-1], degrees[::-1], color=colors[::-1])
plt.title("Top 15 Nodes by Degree")
plt.xlabel("Degree")
plt.tight_layout()
plt.show()
```

## Section 3: Centrality Measures

```python
from graph_tool.all import Graph, betweenness, closeness, pagerank
import pandas as pd

# Create directed graph-tool graph
G_gt = Graph(directed=True)

# Map node names to graph-tool vertices
node_map = {}
for node in pd.concat([filtered_df["x_name"], filtered_df["y_name"]]).unique():
    node_map[node] = G_gt.add_vertex()

# Add directed edges
for _, row in filtered_df.iterrows():
    G_gt.add_edge(node_map[row["x_name"]], node_map[row["y_name"]])

# Centrality measures
degree_centrality = G_gt.degree_property_map("total")
betweenness_centrality, _ = betweenness(G_gt)
closeness_centrality = closeness(G_gt)
pagerank_centrality = pagerank(G_gt)

# Assemble DataFrame
centrality_df = pd.DataFrame({
    "node": list(node_map.keys()),
    "degree_centrality": [degree_centrality[v] for v in G_gt.vertices()],
    "betweenness_centrality": [betweenness_centrality[v] for v in G_gt.vertices()],
    "closeness_centrality": [closeness_centrality[v] for v in G_gt.vertices()],
    "pagerank": [pagerank_centrality[v] for v in G_gt.vertices()]
})

# Save to CSV
centrality_df.to_csv("graph_tool_centrality.csv", index=False)
```

```python
from pyvis.network import Network
from IPython.display import IFrame

def visualize_centrality(df, centrality_df, centrality_type, filename):
    """
    Visualizes a sampled subgraph with node colors based on a selected centrality measure.

    Parameters:
    - df: DataFrame with edges (filtered_primekg.csv)
    - centrality_df: DataFrame with centrality values (graph_tool_centrality.csv)
    - centrality_type: One of ['degree_centrality', 'betweenness_centrality', 'closeness_centrality', 'pagerank']
    - filename: Output HTML filename
    """

    # Sample a subset for visualization
    sample_df = df.sample(n=1000, random_state=42)

    # Valid centrality types
    valid_centralities = ["degree_centrality", "betweenness_centrality", "closeness_centrality", "pagerank"]
    if centrality_type not in valid_centralities:
        raise ValueError(f"Invalid centrality type. Choose from: {valid_centralities}")

    # Initialize PyVis network
    net = Network(height="700px", width="100%", notebook=True, cdn_resources="remote")
    net.barnes_hut()

    # Create lookup dict for fast access
    centrality_lookup = centrality_df.set_index("node")[centrality_type].to_dict()

    # Get quantiles for adaptive coloring
    values = list(centrality_lookup.values())
    q_high = pd.Series(values).quantile(0.90)
    q_mid = pd.Series(values).quantile(0.60)

    # Define color scheme per metric
    color_schemes = {
        "degree_centrality": ("#008080", "#90EE90", "#000080"),  # Teal, Light Green, Navy
        "betweenness_centrality": ("#800080", "#E6E6FA", "#ADD8E6"),  # Purple, Lavender, Light Blue
        "closeness_centrality": ("#4682B4", "#87CEEB", "#D3D3D3"),  # Steel Blue, Sky Blue, Light Gray
        "pagerank": ("#B22222", "#FFB6C1", "#F5F5F5")  # Firebrick, Light Pink, White Smoke
    }

    high_color, mid_color, low_color = color_schemes[centrality_type]

    # Add nodes and edges
    added_nodes = set()

    for _, row in sample_df.iterrows():
        node1, node2 = row["x_name"], row["y_name"]

        for node in [node1, node2]:
            if node in centrality_lookup and node not in added_nodes:
                val = centrality_lookup[node]
                color = (
                    high_color if val >= q_high else
                    mid_color if val >= q_mid else
                    low_color
                )
                net.add_node(
                    node,
                    label=node,
                    title=f"{centrality_type}: {val:.4f}",
                    color=color,
                    size=25,
                    font={"size": 20}
                )
                added_nodes.add(node)

        # Add edge (even if one of the nodes doesn't have centrality)
        net.add_edge(node1, node2)

    # Layout options
    net.set_options("""
    var options = {
      "nodes": {
        "shape": "dot",
        "scaling": { "min": 10, "max": 40 }
      },
      "edges": {
        "color": { "inherit": true },
        "smooth": false
      },
      "physics": {
        "solver": "forceAtlas2Based",
        "forceAtlas2Based": {
          "gravitationalConstant": -50,
          "centralGravity": 0.01,
          "springLength": 100,
          "springConstant": 0.08
        },
        "minVelocity": 0.75
      }
    }
    """)
    print(f"✅ Graph visualization saved as: {filename}")
    net.show(filename)
    IFrame(filename, width="100%", height="750px")
    
```

## Subgraph Visualization

```python
from pyvis.network import Network
import pandas as pd
import networkx as nx
from IPython.display import IFrame, display
from sklearn.preprocessing import MinMaxScaler

def visualize_centrality_subgraph(
    edge_df,
    centrality_df,
    centrality_type,
    top_k=15,
    max_neighbors=10,
    filename="centrality_subgraph.html"
):
    """
    Visualizes a subgraph of top-k nodes by a selected centrality measure and their limited 1-hop neighbors.

    Parameters:
    - edge_df: DataFrame with edges (filtered_df)
    - centrality_df: DataFrame with centrality scores (graph_tool_centrality.csv)
    - centrality_type: Column in centrality_df (e.g. 'pagerank')
    - top_k: Number of top nodes to select
    - max_neighbors: Max 1-hop neighbors per top-k node
    - filename: Output HTML file name
    """

    if centrality_type not in centrality_df.columns:
        raise ValueError(f"{centrality_type} not found in centrality_df columns.")

    # Step 1: Build node type map
    type_map = pd.concat([
        edge_df[["x_name", "x_type"]].rename(columns={"x_name": "node", "x_type": "type"}),
        edge_df[["y_name", "y_type"]].rename(columns={"y_name": "node", "y_type": "type"})
    ]).drop_duplicates().set_index("node")["type"].to_dict()

    # Step 2: Top-k nodes by centrality
    top_nodes = centrality_df.sort_values(centrality_type, ascending=False).head(top_k)["node"].tolist()

    # Step 3: Limit 1-hop neighbors
    def get_limited_subedges(edge_df, top_nodes, max_neighbors=10):
        rows = []
        for node in top_nodes:
            node_edges = edge_df[(edge_df["x_name"] == node) | (edge_df["y_name"] == node)]
            rows.append(node_edges.sample(n=min(max_neighbors, len(node_edges)), random_state=42))
        return pd.concat(rows).drop_duplicates()

    sub_edges = get_limited_subedges(edge_df, top_nodes, max_neighbors=max_neighbors)
    G = nx.from_pandas_edgelist(sub_edges, source="x_name", target="y_name")

    # Step 4: Centrality & normalized lookup
    cent_lookup = centrality_df.set_index("node")[centrality_type].to_dict()

    scaler = MinMaxScaler()
    cent_series = pd.Series(cent_lookup).fillna(0)
    scaled_scores = scaler.fit_transform(cent_series.values.reshape(-1, 1)).flatten()
    scaled_lookup = dict(zip(cent_series.index, scaled_scores))

    # Step 5: PyVis setup
    net = Network(height="750px", width="100%", notebook=True, bgcolor="#ffffff", font_color="black")
    net.force_atlas_2based()

    # Node color mapping
    color_map = {
        "gene/protein": "skyblue",
        "disease": "salmon",
        "effect/phenotype": "orchid",
        "biological_process": "gray"
    }

    # Step 6: Add nodes
    for node in G.nodes():
        score = cent_lookup.get(node, 0)
        scaled_score = scaled_lookup.get(node, 0)
        ntype = type_map.get(node, "unknown")
        color = color_map.get(ntype, "lightgray")

        net.add_node(
            node,
            label=node,
            title=f"{centrality_type}: {score:.5f} | Type: {ntype}",
            size=20 + (scaled_score * 60),
            color=color,
            font={"size": 25}
        )

    # Step 7: Add edges
    for u, v in G.edges():
        net.add_edge(u, v)

    # Step 8: Layout config
    net.set_options("""
    var options = {
      "nodes": {
        "shape": "dot",
        "scaling": { "min": 10, "max": 40 }
      },
      "edges": {
        "color": { "inherit": true },
        "smooth": true
      },
      "physics": {
        "solver": "repulsion",
        "repulsion": {
          "nodeDistance": 150,
          "centralGravity": 0.1
        }
      }
    }
    """)

    net.show(filename)
    display(IFrame(filename, width="100%", height="750px"))

```

## Section 4: Network Topology & Properties of Connected Components

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from networkx.generators.random_graphs import watts_strogatz_graph

def analyze_network_topology(df, relation_type="disease_protein", visualize=True):
    """
    Analyze topology for a specific relation subgraph from a filtered KG edge list.

    Parameters:
    - df: DataFrame with 'relation', 'x_name', 'y_name'
    - relation_type: string (e.g., 'disease_protein')
    - visualize: bool, whether to plot the largest component
    """

    print(f"\n Analyzing relation: '{relation_type}'")

    # Step 1: Filter edges
    sub_df = df[df["relation"] == relation_type].copy()
    if sub_df.empty:
        print(" No edges found for this relation.")
        return

    # Step 2: Create undirected graph
    G = nx.from_pandas_edgelist(sub_df, source="x_name", target="y_name", create_using=nx.Graph())
    print(f" Nodes: {G.number_of_nodes()}, Edges: {G.number_of_edges()}")

    # Step 3: Connected components
    components = list(nx.connected_components(G))
    components = sorted(components, key=len, reverse=True)
    print(f"Connected Components: {len(components)}")

    # Largest component
    G_lcc = G.subgraph(components[0]).copy()
    print(f" Largest Component → Nodes: {G_lcc.number_of_nodes()}, Edges: {G_lcc.number_of_edges()}")

    # Step 4: Clustering
    clustering = nx.average_clustering(G_lcc)
    print(f" Avg Clustering Coefficient (LCC): {clustering:.4f}")

    # Step 5: Shortest Path Length
    try:
        path_length = nx.average_shortest_path_length(G_lcc)
        print(f" Avg Shortest Path Length (LCC): {path_length:.4f}")
    except:
        print(" Cannot compute path length (disconnected component?)")
        path_length = None

    # Step 6: Compare with Watts-Strogatz random graph
    n_nodes = G_lcc.number_of_nodes()
    k = int(sum(dict(G_lcc.degree()).values()) / n_nodes)
    p = 0.05

    try:
        ws_graph = watts_strogatz_graph(n=n_nodes, k=k, p=p)
        ws_clustering = nx.average_clustering(ws_graph)
        ws_path = nx.average_shortest_path_length(ws_graph)

        print(f"\n⚡ Watts-Strogatz Model (n={n_nodes}, k={k}, p={p})")
        print(f"   ↳ Clustering: {ws_clustering:.4f}")
        print(f"   ↳ Avg Path Length: {ws_path:.4f}")

        # Optional check
        if clustering > ws_clustering and (path_length is not None and path_length <= ws_path):
            print(" This network may exhibit small-world properties!")
        else:
            print(" Small-world property not clearly observed.")

    except Exception as e:
        print(f" Error simulating Watts-Strogatz graph: {e}")

    # Step 7: Visualize Largest Component
    if visualize:
        plt.figure(figsize=(10, 10))
        pos = nx.spring_layout(G_lcc, seed=42)
        nx.draw(
            G_lcc, pos,
            node_size=25,
            edge_color="lightgray",
            node_color="salmon",
            with_labels=False,
            alpha=0.7
        )
        plt.title(f"Largest Component: {relation_type} network", fontsize=14)
        plt.show()
```

```python
analyze_network_topology(filtered_df, relation_type="disease_protein")
```

## Section 5: Community Detection

```python
from community import community_louvain
from pyvis.network import Network
from IPython.display import IFrame, display
import random

def detect_and_visualize_communities(
    df,
    relation_types=None,
    max_nodes=1500,
    filename="community_visualization.html"
):
    """
    Detects communities in a multi-relational undirected graph and visualizes them using PyVis.

    Parameters:
    - df: DataFrame with columns ['relation', 'x_name', 'x_type', 'y_name', 'y_type']
    - relation_types: list of relation types to include (default selects key biomedical ones)
    - max_nodes: max number of nodes to include in visualization (for readability)
    - filename: output HTML file for PyVis visualization
    """

    if relation_types is None:
        relation_types = [
            "disease_protein",
            "protein_protein",
            "drug_protein",
            "drug_effect",
            "disease_disease"
        ]

    print("→ Filtering for selected relation types...")
    filtered = df[df["relation"].isin(relation_types)].copy()

    print(f"→ Building graph with {len(filtered)} edges...")
    G = nx.from_pandas_edgelist(
        filtered, 
        source="x_name", target="y_name", 
        create_using=nx.Graph()
    )

    print("→ Detecting communities using Louvain...")
    partition = community_louvain.best_partition(G)
    nx.set_node_attributes(G, partition, "community")

    modularity = community_louvain.modularity(partition, G)
    print(f"Modularity Score: {modularity:.4f}")
    print(f"Total Communities Detected: {len(set(partition.values()))}")

    print("→ Preparing visualization...")
    # Sample subgraph to reduce clutter
    sampled_nodes = random.sample(list(G.nodes()), min(max_nodes, len(G.nodes())))
    G_sub = G.subgraph(sampled_nodes).copy()

    # Initialize PyVis network
    net = Network(height="750px", width="100%", notebook=True, bgcolor="#ffffff", font_color="black")
    net.force_atlas_2based()

    # Build type map for color/shape (optional)
    type_map = pd.concat([
        filtered[["x_name", "x_type"]].rename(columns={"x_name": "node", "x_type": "type"}),
        filtered[["y_name", "y_type"]].rename(columns={"y_name": "node", "y_type": "type"})
    ]).drop_duplicates().set_index("node")["type"].to_dict()

    # Assign distinct colors for communities
    community_colors = {}
    for i, comm in enumerate(set(partition.values())):
        r = lambda: random.randint(100, 255)
        community_colors[comm] = f'rgb({r()},{r()},{r()})'

    for node in G_sub.nodes():
        comm = partition.get(node, -1)
        color = community_colors.get(comm, "gray")
        ntype = type_map.get(node, "unknown")

        net.add_node(
            node,
            label=node,
            color=color,
            title=f"Type: {ntype}, Community: {comm}",
            size=15,
            font={"size": 20}
        )

    for u, v in G_sub.edges():
        net.add_edge(u, v)

    net.set_options("""
    var options = {
      "nodes": {
        "shape": "dot",
        "scaling": { "min": 10, "max": 30 }
      },
      "edges": {
        "color": { "inherit": true },
        "smooth": true
      },
      "physics": {
        "solver": "repulsion",
        "repulsion": {
          "nodeDistance": 150,
          "centralGravity": 0.05
        }
      }
    }
    """)

    net.show(filename)
    display(IFrame(filename, width="100%", height="750px"))
```

```python
from community import community_louvain
from pyvis.network import Network
from IPython.display import IFrame, display
import random
from collections import Counter

def detect_and_visualize_top_communities(
    df,
    relation_types=None,
    max_nodes=1500,
    top_k_communities=10,
    filename="top_communities_visualization.html"
):
    """
    Detects Louvain communities from selected relation types and visualizes only the top K communities.

    Parameters:
    - df: DataFrame with KG edges
    - relation_types: List of relation types to include (optional)
    - max_nodes: Max nodes for sampled subgraph
    - top_k_communities: Only the top K largest communities will be visualized
    - filename: Output HTML filename
    """

    if relation_types is None:
        relation_types = [
            "disease_protein",
            "protein_protein",
            "drug_protein",
            "drug_effect",
            "disease_disease"
        ]

    filtered = df[df["relation"].isin(relation_types)].copy()
    print(f"→ Filtered to {len(filtered)} edges from selected relation types.")

    G = nx.from_pandas_edgelist(filtered, source="x_name", target="y_name", create_using=nx.Graph())

    print("→ Running Louvain community detection...")
    partition = community_louvain.best_partition(G)
    nx.set_node_attributes(G, partition, "community")

    modularity = community_louvain.modularity(partition, G)
    print(f"→ Modularity Score: {modularity:.4f}")

    # Count community sizes
    community_counts = Counter(partition.values())
    top_communities = set([c for c, _ in community_counts.most_common(top_k_communities)])
    print(f"→ Visualizing Top {top_k_communities} communities (out of {len(community_counts)} total)")

    # Filter nodes for only top communities
    selected_nodes = [node for node, comm in partition.items() if comm in top_communities]
    G_sub = G.subgraph(selected_nodes).copy()

    # Sample for manageable size
    if len(G_sub) > max_nodes:
        G_sub = G_sub.subgraph(random.sample(list(G_sub.nodes()), max_nodes)).copy()

    # Recalculate visible partition for subgraph
    partition_sub = {n: partition[n] for n in G_sub.nodes()}

    # Build node type map
    type_map = pd.concat([
        filtered[["x_name", "x_type"]].rename(columns={"x_name": "node", "x_type": "type"}),
        filtered[["y_name", "y_type"]].rename(columns={"y_name": "node", "y_type": "type"})
    ]).drop_duplicates().set_index("node")["type"].to_dict()

    # Assign color per community
    community_colors = {}
    for c in top_communities:
        r = lambda: random.randint(100, 255)
        community_colors[c] = f'rgb({r()},{r()},{r()})'

    # Shape mapping for node types
    shape_map = {
        "gene/protein": "dot",
        "disease": "triangle",
        "drug": "box",
        "effect/phenotype": "diamond",
        "biological_process": "star"
    }

    # Create PyVis network
    net = Network(height="750px", width="100%", notebook=True, cdn_resources="remote")
    net.force_atlas_2based()

    for node in G_sub.nodes():
        comm = partition_sub.get(node, -1)
        color = community_colors.get(comm, "gray")
        ntype = type_map.get(node, "unknown")
        shape = shape_map.get(ntype, "dot")

        net.add_node(
            node,
            label=node,
            color=color,
            shape=shape,
            title=f"Type: {ntype}, Community: {comm}",
            size=18,
            font={"size": 20}
        )

    for u, v in G_sub.edges():
        net.add_edge(u, v)

    # Apply layout and show
    net.set_options("""
    var options = {
      "nodes": {
        "scaling": { "min": 10, "max": 30 }
      },
      "edges": {
        "color": { "inherit": true },
        "smooth": true
      },
      "physics": {
        "solver": "repulsion",
        "repulsion": {
          "nodeDistance": 150,
          "centralGravity": 0.05
        }
      }
    }
    """)

    net.show(filename)
    display(IFrame(filename, width="100%", height="750px"))
```

## Section 6: Semantic & Structural Subgraphs

```python
import pandas as pd
import networkx as nx
from pyvis.network import Network
from IPython.display import display, IFrame

def visualize_khop_subgraph(
    df,
    target_node,
    k=2,
    max_nodes=150,
    filename="khop_subgraph.html"
):
    """
    Visualize a k-hop semantic subgraph around a target node with node and edge coloring.

    Parameters:
    - df: DataFrame containing edges with columns ['x_name', 'x_type', 'y_name', 'y_type', 'relation']
    - target_node: Name of the node to center the subgraph on
    - k: Number of hops (default: 2)
    - max_nodes: Max number of nodes to visualize (to keep it readable)
    - filename: Output HTML file for PyVis visualization
    """

    print(f"→ Building full undirected graph for traversal...")
    G = nx.from_pandas_edgelist(df, source="x_name", target="y_name")

    if target_node not in G.nodes():
        print(f"Node '{target_node}' not found in graph.")
        return

    print(f"→ Extracting {k}-hop neighborhood of '{target_node}'")
    nodes_khop = nx.single_source_shortest_path_length(G, target_node, cutoff=k).keys()
    subgraph = G.subgraph(nodes_khop).copy()

    if len(subgraph) > max_nodes:
        subgraph = subgraph.subgraph(list(subgraph.nodes())[:max_nodes])
        print(f"→ Subgraph trimmed to {max_nodes} nodes for clarity.")

    # Build node and edge metadata
    type_map = pd.concat([
        df[["x_name", "x_type"]].rename(columns={"x_name": "node", "x_type": "type"}),
        df[["y_name", "y_type"]].rename(columns={"y_name": "node", "y_type": "type"})
    ]).drop_duplicates().set_index("node")["type"].to_dict()

    edge_type_lookup = {
        (row["x_name"], row["y_name"]): row["relation"]
        for _, row in df.iterrows()
    }
    edge_type_lookup.update({
        (row["y_name"], row["x_name"]): row["relation"]
        for _, row in df.iterrows()
    })

    # Color maps
    node_color_map = {
        "gene/protein": "skyblue",
        "disease": "salmon",
        "drug": "lightgreen",
        "effect/phenotype": "orchid",
        "biological_process": "gray"
    }

    edge_color_map = {
        "protein_protein": "gray",
        "disease_protein": "purple",
        "drug_protein": "green",
        "drug_effect": "orange",
        "disease_disease": "red",
        "bioprocess_protein": "blue"
    }

    # PyVis setup
    net = Network(height="750px", width="100%", notebook=True, cdn_resources="remote")
    net.force_atlas_2based()

    for node in subgraph.nodes():
        ntype = type_map.get(node, "unknown")
        color = node_color_map.get(ntype, "lightgray")
        net.add_node(
            node,
            label=node,
            title=f"{node} ({ntype})",
            color=color,
            shape="dot",
            size=25 if node == target_node else 15,
            font={"size": 22 if node == target_node else 16}
        )

    for u, v in subgraph.edges():
        rel = edge_type_lookup.get((u, v), "unknown")
        edge_color = edge_color_map.get(rel, "lightgray")
        net.add_edge(u, v, title=rel, color=edge_color, width=2)

    net.set_options("""
    var options = {
      "nodes": {
        "scaling": { "min": 10, "max": 30 }
      },
      "edges": {
        "smooth": true
      },
      "physics": {
        "solver": "forceAtlas2Based",
        "forceAtlas2Based": {
          "gravitationalConstant": -50,
          "centralGravity": 0.02
        }
      }
    }
    """)

    net.show(filename)
    display(IFrame(filename, width="100%", height="750px"))
```

## Structured Khop subgraph with Edges

```python
import pandas as pd
import networkx as nx
from pyvis.network import Network
from IPython.display import IFrame, display

def visualize_structured_khop_subgraph(
    df,
    target_node,
    k=2,
    max_edges=200,
    filename="structured_khop_subgraph_v2.html"
):
    """
    Extracts a k-hop subgraph around a node and visualizes nodes and typed edges using PyVis.

    Parameters:
    - df: DataFrame with 'x_name', 'y_name', 'relation', 'x_type', 'y_type'
    - target_node: Central node to extract the subgraph from
    - k: Number of hops to consider
    - max_edges: Trim subgraph if it's too large
    - filename: Output HTML filename
    """

    # Step 1: Build undirected graph for neighbor traversal
    print(f"Finding {k}-hop neighbors of '{target_node}'")
    G_full = nx.from_pandas_edgelist(df, source="x_name", target="y_name")

    if target_node not in G_full.nodes:
        print(f"Target node '{target_node}' not found.")
        return

    khop_nodes = set(nx.single_source_shortest_path_length(G_full, target_node, cutoff=k).keys())

    # Step 2: Filter the dataframe to only keep edges within khop nodes
    sub_df = df[
        df["x_name"].isin(khop_nodes) & df["y_name"].isin(khop_nodes)
    ].copy()

    print(f" Subgraph contains {len(sub_df)} edges.")

    if len(sub_df) > max_edges:
        sub_df = sub_df.sample(n=max_edges, random_state=42)
        print(f" Subgraph trimmed to {max_edges} edges.")

    # Step 3: Remove nodes not part of any remaining edge
    active_nodes = set(sub_df["x_name"]).union(sub_df["y_name"])

    # Step 4: Node & edge color maps
    node_color_map = {
        "gene/protein": "skyblue",
        "disease": "salmon",
        "drug": "lightgreen",
        "effect/phenotype": "orchid",
        "biological_process": "gray"
    }

    edge_color_map = {
        "protein_protein": "gray",
        "disease_protein": "purple",
        "drug_protein": "green",
        "drug_effect": "orange",
        "disease_disease": "red",
        "bioprocess_protein": "blue"
    }

    # Step 5: Init PyVis
    net = Network(height="750px", width="100%", notebook=True, cdn_resources="remote")
    net.force_atlas_2based()

    # Add nodes
    for node in active_nodes:
        node_type = None
        try:
            node_type = df.loc[df["x_name"] == node, "x_type"].values[0]
        except:
            try:
                node_type = df.loc[df["y_name"] == node, "y_type"].values[0]
            except:
                node_type = "unknown"

        color = node_color_map.get(node_type, "lightgray")
        label = node if len(node) <= 25 else node[:22] + "..."

        net.add_node(
            node,
            label=label,
            title=f"{node} ({node_type})",
            color=color,
            size=25 if node == target_node else 15,
            font={"size": 20 if node == target_node else 16}
        )

    # Add edges with relation labels
    for _, row in sub_df.iterrows():
        rel = row["relation"]
        color = edge_color_map.get(rel, "lightgray")

        net.add_edge(
            row["x_name"],
            row["y_name"],
            title=rel,
            color=color,
            width=2
        )

    # Optional layout config
    net.set_options("""
    var options = {
      "nodes": {
        "scaling": { "min": 10, "max": 30 }
      },
      "edges": {
        "smooth": true
      },
      "physics": {
        "solver": "forceAtlas2Based",
        "forceAtlas2Based": {
          "gravitationalConstant": -50,
          "centralGravity": 0.01
        }
      }
    }
    """)

    # Render
    net.show(filename)
    display(IFrame(filename, width="100%", height="750px"))
```

## Graph Traversals & Explanatory Views Shortest Paths

```python
import pandas as pd
import networkx as nx
from pyvis.network import Network
from IPython.display import display, IFrame

def visualize_shortest_path(df, source, target, filename="shortest_path.html"):
    """
    Find and visualize the shortest path between two nodes using PyVis.
    
    Parameters:
    - df: DataFrame with 'x_name', 'y_name', 'relation', 'x_type', 'y_type'
    - source: Starting node (e.g. a drug or disease)
    - target: Destination node (e.g. a phenotype or gene)
    - filename: HTML output file
    """

    # Step 1: Create undirected graph
    G = nx.from_pandas_edgelist(df, source="x_name", target="y_name")

    if source not in G or target not in G:
        print(f"Either '{source}' or '{target}' not found in graph.")
        return

    try:
        path = nx.shortest_path(G, source=source, target=target)
        print(f"→ Shortest path from '{source}' to '{target}' found. Length = {len(path)-1}")
    except nx.NetworkXNoPath:
        print(f"No path exists between '{source}' and '{target}'")
        return

    # Step 2: Extract subgraph edges between consecutive pairs in path
    path_edges = list(zip(path[:-1], path[1:]))

    # Filter original df to find matching edges with relation types
    rel_edges = []
    for u, v in path_edges:
        row = df[((df["x_name"] == u) & (df["y_name"] == v)) | ((df["x_name"] == v) & (df["y_name"] == u))]
        if not row.empty:
            rel_edges.append(row.iloc[0])
        else:
            rel_edges.append(pd.Series({"x_name": u, "y_name": v, "relation": "unknown"}))

    path_df = pd.DataFrame(rel_edges)

    # Step 3: Get node types
    type_map = pd.concat([
        df[["x_name", "x_type"]].rename(columns={"x_name": "node", "x_type": "type"}),
        df[["y_name", "y_type"]].rename(columns={"y_name": "node", "y_type": "type"})
    ]).drop_duplicates().set_index("node")["type"].to_dict()

    # Step 4: Color maps
    node_color_map = {
        "gene/protein": "skyblue",
        "disease": "salmon",
        "drug": "lightgreen",
        "effect/phenotype": "orchid",
        "biological_process": "gray"
    }

    edge_color_map = {
        "protein_protein": "gray",
        "disease_protein": "purple",
        "drug_protein": "green",
        "drug_effect": "orange",
        "disease_disease": "red",
        "bioprocess_protein": "blue"
    }

    # Step 5: PyVis
    net = Network(height="650px", width="100%", notebook=True, cdn_resources="remote")
    net.force_atlas_2based()

    for node in path:
        ntype = type_map.get(node, "unknown")
        color = node_color_map.get(ntype, "lightgray")
        net.add_node(
            node,
            label=node if len(node) <= 25 else node[:22] + "...",
            title=f"{node} ({ntype})",
            color=color,
            size=25,
            font={"size": 20}
        )

    for _, row in path_df.iterrows():
        rel = row["relation"]
        color = edge_color_map.get(rel, "lightgray")
        net.add_edge(
            row["x_name"],
            row["y_name"],
            title=rel,
            label=rel,
            color=color,
            width=3,
            font={"size": 16, "align": "top"}
        )

    net.set_options("""
    var options = {
      "nodes": {
        "shape": "dot",
        "scaling": { "min": 10, "max": 30 }
      },
      "edges": {
        "smooth": true
      },
      "physics": {
        "solver": "forceAtlas2Based",
        "forceAtlas2Based": {
          "gravitationalConstant": -60,
          "centralGravity": 0.01
        }
      }
    }
    """)

    net.show(filename)
    display(IFrame(filename, width="100%", height="700px"))
```

```python
def visualize_directed_shortest_path(df, source, target, filename="shortest_directed_path.html"):
    """
    Visualizes the shortest directed path between two nodes using PyVis.

    Parameters:
    - df: DataFrame with x_name, y_name, relation, x_type, y_type
    - source: starting node name
    - target: ending node name
    - filename: output HTML path
    """

    # Step 1: Build directed graph
    G = nx.from_pandas_edgelist(df, source="x_name", target="y_name", create_using=nx.DiGraph())

    if source not in G or target not in G:
        print(f"Either '{source}' or '{target}' not found in the graph.")
        return

    try:
        path = nx.shortest_path(G, source=source, target=target)
        print(f"→ Directed path from '{source}' to '{target}' found. Length = {len(path)-1}")
    except nx.NetworkXNoPath:
        print(f"No directed path exists between '{source}' and '{target}'.")
        return

    # Step 2: Retrieve edge metadata from df
    edges_in_path = list(zip(path[:-1], path[1:]))
    rel_edges = []
    for u, v in edges_in_path:
        row = df[(df["x_name"] == u) & (df["y_name"] == v)]
        if not row.empty:
            rel_edges.append(row.iloc[0])
        else:
            rel_edges.append(pd.Series({"x_name": u, "y_name": v, "relation": "unknown"}))

    path_df = pd.DataFrame(rel_edges)

    # Step 3: Node types
    type_map = pd.concat([
        df[["x_name", "x_type"]].rename(columns={"x_name": "node", "x_type": "type"}),
        df[["y_name", "y_type"]].rename(columns={"y_name": "node", "y_type": "type"})
    ]).drop_duplicates().set_index("node")["type"].to_dict()

    # Step 4: Color maps
    node_color_map = {
        "gene/protein": "skyblue",
        "disease": "salmon",
        "drug": "lightgreen",
        "effect/phenotype": "orchid",
        "biological_process": "gray"
    }

    edge_color_map = {
        "protein_protein": "gray",
        "disease_protein": "purple",
        "drug_protein": "green",
        "drug_effect": "orange",
        "disease_disease": "red",
        "bioprocess_protein": "blue"
    }

    # Step 5: PyVis Visualization
    net = Network(height="700px", width="100%", notebook=True, directed=True, cdn_resources="remote")
    net.force_atlas_2based()

    for node in path:
        ntype = type_map.get(node, "unknown")
        color = node_color_map.get(ntype, "lightgray")
        label = node if len(node) <= 25 else node[:22] + "..."
        net.add_node(
            node,
            label=label,
            title=f"{node} ({ntype})",
            color=color,
            size=35,
            font={"size": 24}
        )

    for _, row in path_df.iterrows():
        rel = row["relation"]
        color = edge_color_map.get(rel, "lightgray")
        net.add_edge(
            row["x_name"],
            row["y_name"],
            title=rel,
            color=color,
            width=3,
            arrows="to",
            font={"size": 18, "align": "top"}
        )

    # Step 6: Layout options (more spacing between nodes)
    net.set_options("""
    var options = {
      "nodes": {
        "scaling": { "min": 15, "max": 40 }
      },
      "edges": {
        "smooth": true,
        "length": 400
      },
      "physics": {
        "solver": "forceAtlas2Based",
        "forceAtlas2Based": {
          "gravitationalConstant": -50,
          "centralGravity": 0.01,
          "springLength": 300,
          "springConstant": 0.02
        },
        "minVelocity": 0.75
      }
    }
    """)

    net.show(filename)
    display(IFrame(filename, width="100%", height="750px"))
```

## Exploring Causality

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
from pyvis.network import Network
from IPython.display import IFrame, display
from causallearn.search.ConstraintBased.PC import pc
from causallearn.utils.GraphUtils import GraphUtils
from causallearn.utils.cit import fisherz
import numpy as np

# Optional: install causal discovery packages (run in notebook if needed)
# !pip install causallearn


def analyze_causal_subgraph(df, display_relation_column="display_relation", k_hop_center=None, k_hop=2, html_filename="causal_subgraph.html"):
    """
    Analyze and visualize causal-like subgraphs in PrimeKG.

    Parameters:
    - df: Full filtered_df DataFrame.
    - display_relation_column: Column that stores display relation labels.
    - k_hop_center: Optional node name to extract a k-hop neighborhood for zoom-in.
    - k_hop: Number of hops for k-hop neighborhood.
    - html_filename: Output filename for PyVis visualization.
    """
    # Step 1: Define causal relations
    causal_labels = [
        "side effect", "target", "indication",
        "contraindication", "enzyme", "transporter", "carrier"
    ]

    # Step 2: Extract causal edges
    causal_df = df[df[display_relation_column].isin(causal_labels)].copy()
    print(f"Total causal edges extracted: {len(causal_df):,}")

    # Step 3: Node type summary
    node_types = pd.concat([
        causal_df[["x_name", "x_type"]].rename(columns={"x_name": "node", "x_type": "type"}),
        causal_df[["y_name", "y_type"]].rename(columns={"y_name": "node", "y_type": "type"})
    ])
    type_counts = node_types.drop_duplicates().groupby("type").size().reset_index(name="count")
    print("\nNode Type Distribution:")
    print(type_counts.sort_values("count", ascending=False))

    # Optional: plot type distribution
    plt.figure(figsize=(8, 4))
    sns.barplot(data=type_counts, x="type", y="count", palette="viridis")
    plt.xticks(rotation=45)
    plt.title("Node Type Distribution in Causal Subgraph")
    plt.tight_layout()
    plt.show()

    # Step 4: Build directed graph
    G = nx.from_pandas_edgelist(causal_df, source="x_name", target="y_name", create_using=nx.DiGraph())

    # Optional: k-hop neighborhood
    if k_hop_center and k_hop_center in G:
        nodes = nx.single_source_shortest_path_length(G, k_hop_center, cutoff=k_hop).keys()
        G = G.subgraph(nodes).copy()
        print(f"Extracted {len(G.nodes())} nodes in {k_hop}-hop neighborhood of '{k_hop_center}'.")

    # Step 5: Visualization with PyVis
    type_map = node_types.drop_duplicates().set_index("node")["type"].to_dict()
    relation_map = causal_df.set_index(["x_name", "y_name"])[display_relation_column].to_dict()

    color_map = {
        "gene/protein": "skyblue",
        "disease": "salmon",
        "drug": "lightgreen",
        "effect/phenotype": "orchid",
        "biological_process": "gray"
    }

    net = Network(height="700px", width="100%", directed=True, notebook=True, cdn_resources="remote")
    net.force_atlas_2based()

    for node in G.nodes():
        ntype = type_map.get(node, "unknown")
        color = color_map.get(ntype, "lightgray")
        label = node if len(node) <= 25 else node[:22] + "..."
        net.add_node(node, label=label, title=f"{node} ({ntype})", color=color, size=25)

    for u, v in G.edges():
        rel = relation_map.get((u, v), "causal")
        net.add_edge(u, v, title=rel, label=rel, color="purple", arrows="to")

    net.set_options("""
    var options = {
      "edges": {
        "smooth": true,
        "length": 400
      },
      "physics": {
        "solver": "forceAtlas2Based",
        "forceAtlas2Based": {
          "gravitationalConstant": -50,
          "centralGravity": 0.01,
          "springLength": 300,
          "springConstant": 0.02
        },
        "minVelocity": 0.75
      }
    }
    """)

    net.show(html_filename)
    display(IFrame(html_filename, width="100%", height="750px"))

    return G  # Useful for next step: causal discovery


def run_pc_causal_discovery(G):
    """
    Runs the PC causal discovery algorithm on a graph with binary adjacency matrix.

    Parameters:
    - G: networkx.DiGraph or Graph with nodes of interest

    Returns:
    - Causal graph learned by the PC algorithm
    """
    print("\nRunning Causal Discovery (PC algorithm)...")

    nodes = list(G.nodes())
    node_index = {node: idx for idx, node in enumerate(nodes)}
    A = nx.to_numpy_array(G, nodelist=nodes)  # binary adjacency matrix (0/1)

    # Simulate features: use adjacency rows as pseudo-samples
    data = A.copy()  # shape (n_samples, n_features) with binary patterns

    # Run PC
    pc_output = pc(data, fisherz, 0.05)

    learned_graph = GraphUtils.to_nx_graph(pc_output.G, labels=nodes)

    print("Causal graph edges detected:")
    print(list(learned_graph.edges())[:10])
    return learned_graph
```

```python
causal_G = analyze_causal_subgraph(filtered_df, k_hop_center="breast cancer", k_hop=2)
learned_G = run_pc_causal_discovery(causal_G)
```
